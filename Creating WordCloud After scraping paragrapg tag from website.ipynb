# Scrape paragraph

```python
ptags = browser.find_elements_by_css_selector("p") #for more than one paragraph
```

```python
len(ptags)
```

```python
#reading text from paragraph tag from web
ptags_texts = ''
for tag in ptags:
    #print(tag.text)
    text = tag.text
    ptags_texts = ptags_texts +" "+ text
len(ptags_texts)
```

```python
ptags_texts
```

```python
#using list comprehension
ptags_texts = " ".join([tag.text for tag in ptags])
len(ptags_texts)
```

# method 2

```python
#better approach
articles_tags = browser.find_elements_by_css_selector("div.article")

df_articles=pd.DataFrame()

for tag in articles_tags:
    title = tag.find_element_by_css_selector('p[name="title"]').text
    content = tag.find_element_by_css_selector('p[name="content"]').text
    
    row = {"title":title,"content":content}
    df_articles=df_articles.append(row,ignore_index=True)
df_articles
```
