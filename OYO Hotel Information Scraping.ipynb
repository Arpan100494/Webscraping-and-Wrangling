```python
import pandas as pd
from selenium import webdriver
browser = webdriver.Chrome("chromedriver.exe")
```

```python
#Hotel name,location,price,discount,ratings,no.of ratings,image url

url = "https://www.oyorooms.com/oyos-in-bangalore"
browser.get(url)
```

```python
parents = browser.find_elements_by_css_selector('div.newHotelCard')
len(parents)
```

```python
hotel_df = pd.DataFrame()

for hotel in parents:
    try:
        hotel_id = hotel.get_attribute("id")
    except:
        hotel_id = None
    try:
        hotel_name = hotel.find_element_by_css_selector("span.newHotelCard__hotelName").text
    except:
        hotel_name = None
    try:
        hotel_name = hotel.find_element_by_css_selector("span.newHotelCard__hotelName").text   
    except:
        hotel_name = None
    try:
        location = hotel.find_element_by_css_selector("span.newHotelCard__hotelAddress").text
    except:
        location = None
    try:
        discount = hotel.find_element_by_css_selector("div.newHotelCard__discount").text
    except:
        discount = None
    try:
        rating = hotel.find_element_by_css_selector("div.hotelRating__count").text
    except:
        rating = None
    try:
        rating2 = hotel.find_element_by_css_selector("span.hotelRating__value--excellent").text
    except:
        rating2 = None
    try:
        image_url = hotel.find_element_by_css_selector("img.carousal__image").get_attribute("src")
    except:
        image_url = None  
    try:
        price = hotel.find_element_by_css_selector("span.newHotelCard__revisedPricing").text
    except:
        price = None 
        
    df = {"hotel_id":hotel_id,"hotel_name":hotel_name,"price":price,"discount":discount,"rating":rating,
         "rating2":rating2,"location":location,"image_url":image_url}
    #print(hotel_id,hotel_name,location,discount,rating,rating2,image_url,price)
    hotel_df = hotel_df.append(df,ignore_index=True)
```

```python
url = "https://www.naukri.com/data-science-jobs"
browser.get(url)
```

```python
    def get_attribute(parent,selector,attribute):
    try:
        if attribute == "text":
            val = parent.find_element_by_css_selector(selector).text
        else:
            val = parent.find_element_by_css_selector(selector).get_attribute(attribute)
    except:
        val = None
    return val
```

```python
import time
jobs_df = pd.DataFrame()
while True:
    parents = browser.find_elements_by_css_selector('div.row[type="tuple"]')
    
    for job in parents:
        job_dict={
        "id":job.get_attribute("id"),
        "url":job.get_attribute("data-url"),
        "title":get_attribute(job,"li.desig","title"),
        "company":get_attribute(job,"span.org","text"),
        "experience":get_attribute(job,"span.exp","text"),
        "location":get_attribute(job,"span.loc","text"),
        "skills":get_attribute(job,"span.skill","text"),    
        "salary":get_attribute(job,"span.salary","text"), 
        "date":get_attribute(job,"span.date","text")
        }
        jobs_df = jobs_df.append(job_dict,ignore_index=True)
        print(jobs_df.shape)
    try:
        browser.find_element_by_css_selector("button.grayBtn").click()
        time.sleep(2)
    except:
        break
```
