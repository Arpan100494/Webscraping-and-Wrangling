```python
import pandas as pd
```

```python
from selenium import webdriver
browser = webdriver.Chrome("chromedriver.exe")
```

```python
#browser.get("http://localhost:8000/new_book.html")
browser.get("https://www.naukri.com/data-science-jobs")
```

```python
parents = browser.find_elements_by_css_selector('div.row[type="tuple"]')
len(parents)
```

```python
job_df = pd.DataFrame()

for job in parents:
    job_id = job.get_attribute("id")
    url = job.get_attribute("data-url")
    title = job.find_element_by_css_selector("li.desig").get_attribute('title')
    company = job.find_element_by_css_selector("span.org").text
    experience = job.find_element_by_css_selector("span.exp").text
    location = job.find_element_by_css_selector("span.loc").text
    keyskill = job.find_element_by_css_selector("span.skill").text
    salary = job.find_element_by_css_selector("span.salary").text
    date = job.find_element_by_css_selector("span.date").text
    job_dict = {"job_id":job_id,"url":url,"title":title,"company":company,"experience":experience,"location":location,"keyskill":keyskill,
               "salary":salary,"date":date}
    job_df = job_df.append(job_dict,ignore_index=True)
job_df
```

```python
#create a csv file
job_df.to_csv("jobs_datascience.csv",index=False)
```

```python
def get_attribute(parent,selector,attribute):
    try:
        if attribute == "text":
            val = parent.find_element_by_css_selector(selector).text
        else:
            val = parent.find_element_by_css_selector(selector).get_attribute(attribute)
    except:
        val = None
    return val
```

```python
import time
jobs_df = pd.DataFrame()
for i in range(1,10):
    url = "https://www.naukri.com/data-science-jobs-%d" % (i)
    browser.get(url)
    time.sleep(2)
    print(i)
    parents = browser.find_elements_by_css_selector('div.row[type="tuple"]')
    
    for job in parents:
        job_dict={
        "id":job.get_attribute("job_id"),
        "url":job.get_attribute("data-url"),
        "title":get_attribute(job,"li.desig","title"),
        "company":get_attribute(job,"span.org","text"),
        "experience":get_attribute(job,"span.exp","text"),
        "location":get_attribute(job,"span.loc","text"),
        "skills":get_attribute(job,"span.skill","text"),    
        "salary":get_attribute(job,"span.salary","text"), 
        "date":get_attribute(job,"span.date","text")
        }
        jobs_df = jobs_df.append(job_dict,ignore_index=True)
```
