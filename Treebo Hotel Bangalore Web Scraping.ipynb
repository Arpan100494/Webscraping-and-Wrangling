```python
import pandas as pd
from selenium import webdriver
browser = webdriver.Chrome("chromedriver.exe")
```

```python
#hotel_name,address,price,reviews and standard
```

```python
url = "https://www.treebo.com/hotels-in-bangalore"
browser.get(url)
```

```python
parents = browser.find_elements_by_css_selector("div.flex-row")
len(parents)
```

```python
df = pd.DataFrame()

for hotel in parents:
    try:
        hotel_name = hotel.find_element_by_css_selector('span[itemprop="name"]').text
    except:
        hotel_name = None
    try:
        hotel_address = hotel.find_element_by_css_selector('div.text-2').text
    except:
        hotel_address = None
    try:
        hotel_price = hotel.find_element_by_css_selector('div.r-hotel__price').text
    except:
        hotel_price = None
    try:
        hotel_reviews = hotel.find_element_by_css_selector('div.text-3').text
    except:
        hotel_reviews = None
    try:
        hotel_class= hotel.find_element_by_css_selector('div.iwgzbo').text
    except:
        hotel_class = None
        
    dict1 = pd.Series({"hotel_name":hotel_name,"hotel_address":hotel_address,"hotel_price":hotel_price,"hotel_reviews":hotel_reviews,
                      "hotel_class":hotel_class})
    df = df.append(dict1,ignore_index=True)
    #print(hotel_class)
```
